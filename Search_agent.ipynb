```python
import requests
from bs4 import BeautifulSoup, SoupStrainer

def scrape_links(links):
    for url in links:
        try:
            # Fetch the content of the URL
            print(f"Fetching content from: {url}")
            response = requests.get(url, headers={"User-Agent": "Mozilla/5.0"})
            
            # Check if the request was successful
            if response.status_code == 200:
                # Use SoupStrainer to filter specific tags
                strainer = SoupStrainer(['p', 'h1', 'h2', 'h3', 'h4', 'h5', 'h6'])
                soup = BeautifulSoup(response.text, 'html.parser', parse_only=strainer)
                
                # Extract text from the filtered tags
                text_content = [tag.get_text(strip=True) for tag in soup if tag.name in ('p', 'h1', 'h2', 'h3', 'h4', 'h5', 'h6')]
                
                # Combine the extracted text
                combined_text = "\n".join(text_content)
                
                # Save the extracted content to a file
                filename = url.replace('https://', '').replace('http://', '').replace('/', '_') + ".txt"
                with open(f"Docs/{filename}", 'w', encoding='utf-8') as file:
                    file.write(combined_text)
                
                print(f"Text from {url} saved to {filename}\n")
            else:
                print(f"Failed to retrieve {url} (Status code: {response.status_code})")
        
        except Exception as e:
            print(f"An error occurred while processing {url}: {e}")

# List of links to scrape
links = search_links

# Call the function to scrape the links
scrape_links(links)
```

```python
import base64
import mimetypes
import os
from google import genai
from google.genai import types
from pydantic import BaseModel, Field
import json

# Define a Pydantic model for structured output, as discussed previously.
# This ensures a consistent JSON format from the API.
class SimilarCompany(BaseModel):
    """
    A structured representation of a company similar to the one being analyzed.
    """
    company_name: str = Field(..., description="The name of the similar company.")
    reason_for_similarity: str = Field(..., description="A brief explanation of why this company is similar.")
    key_products: list[str] = Field(..., description="A list of key products or services provided by the company.")

class SimilarCompaniesList(BaseModel):
    companies: list[SimilarCompany]

def generate(combined_text_document):
    """
    Generates a list of similar companies using the Gemini API from a combined text document.
    """
    client = genai.Client(
        api_key=os.environ.get("AIzaSyBQ3KMmBij5bpA8KdM_iFF9DKOHJGomUYo"),
    )

    model = "gemini-2.5-flash" 
    
    # The prompt now includes instructions for structured output and uses the input text.
    prompt_text = f"""
    You are a business research assistant. Your task is to analyze the following combined text document, which contains information about a specific company.
    Based on the provided text, identify and list 5 to 6 companies that are similar in terms of business model, industry, and key products.
    
    For each similar company, provide:
    1. The company's name.
    2. A brief reason for the similarity.
    3. A list of its key products or services.

    Provide the output as a JSON object that strictly adheres to the provided schema. Do not include any additional text or formatting.
    
    Combined Text Document:
    ---
    {combined_text_document}
    ---
    """
    
    contents = [
        types.Content(
            role="user",
            parts=[
                types.Part.from_text(text=prompt_text),
            ],
        ),
    ]

    # The generate_content_config should specify the response mime type and schema for JSON output
    generate_content_config = types.GenerateContentConfig(
        response_mime_type="application/json",
        response_schema=SimilarCompaniesList.model_json_schema()
    )

    try:
        # Use a non-streaming call for this request for a single JSON object output
        response = client.models.generate_content(
            model=model,
            contents=contents,
            config=generate_content_config,
        )

        # Assuming the response is successful and contains valid JSON
        if response.text:
            json_output = json.loads(response.text)
            print("Successfully received structured JSON output:")
            print(json.dumps(json_output, indent=2))
            
            # You can now use the Pydantic model to work with the data
            similar_companies_list = SimilarCompaniesList(**json_output)
            print("\nParsed data using Pydantic:")
            for company in similar_companies_list.companies:
                print(f"Company: {company.company_name}")
                print(f"Reason: {company.reason_for_similarity}")
                print(f"Products: {', '.join(company.key_products)}\n")
        else:
            print("No text content in the API response.")

    except Exception as e:
        print(f"An error occurred: {e}")

if __name__ == "__main__":
    generate(combined_text_document=combined_text)
```
